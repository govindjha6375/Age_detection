```python
%matplotlib inline 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import os

from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from keras.models import Model
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
home_dir = "C:/Users/91859/Desktop/age_detection/UTKFace"

full_paths = []
ages = []


# Parse the directory
for file in tqdm(os.listdir(home_dir)):
    age, gender = file.split('_')[:2]
    full_paths.append(os.path.join(home_dir, file))
    ages.append(int(age))
```

```python

data = pd.DataFrame({
    'path': full_paths,
    'age': ages,
    })
data.sample(10)
```

```python
data_sample = data.sample(15).reset_index(drop=True)
       
for i in range(0, 15):
    path = data_sample.loc[i, 'path']
    age = data_sample.loc[i, 'age']
    img = plt.imread(path)
    plt.subplot(3, 5, i+1)
    plt.imshow(img)
    plt.title(age, fontsize=6)
    plt.axis('off')
plt.tight_layout()
```

```python
# Split the data to train, validation and split datasets 
X_train, X_test_val = train_test_split(data, test_size=0.2, random_state=42)

# Split test and val evenly
X_test, X_val = train_test_split(X_test_val, test_size=0.5, random_state=42)
```

```python
# Set the data generators
BATCH=64

# Add some data augmentation
datagen_train = ImageDataGenerator(
    rescale=1/255.,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.7, 1.2],
    fill_mode='nearest'
)
# Datagen for val and test
datagen_val_test = ImageDataGenerator(rescale=1/255.)

train_generator = datagen_train.flow_from_dataframe(
    dataframe=X_train,
    x_col='path',
    y_col=['age'],
    target_size=(128, 128),
    batch_size=BATCH,
    class_mode='raw'
)

val_generator = datagen_val_test.flow_from_dataframe(
    dataframe=X_val,
    x_col='path',
    y_col=['age'],
    target_size=(128, 128),
    batch_size=BATCH,
    class_mode='raw'
)
test_generator = datagen_val_test.flow_from_dataframe(
    dataframe=X_test,
    x_col='path',
    y_col=['age'],
    target_size=(128, 128),
    batch_size=BATCH,
    class_mode='raw',
)
```

```python
from tensorflow.keras.utils import Sequence

class CustomDataGenerator(Sequence):
    def __init__(self, generator):
        self.generator = generator
    
    def __len__(self):
        return len(self.generator)
    
    def __getitem__(self, index):
        images, labels = self.generator[index]
        age_labels = labels
        return images, {'age_out': age_labels}
    
    def on_epoch_end(self):
        self.generator.on_epoch_end()
    
    def __iter__(self):
        for item in (self[i] for i in range(len(self))):
            yield item

train_generator = CustomDataGenerator(train_generator)
val_generator = CustomDataGenerator(val_generator)
test_generator = CustomDataGenerator(test_generator)
```

```python
# The base model - InceptionV3
base = InceptionV3(include_top = False, 
                   weights='imagenet',
                   input_tensor=None, 
                   pooling=None, 
                   input_shape=(128, 128, 3))

for layer in base.layers[:136]:
    layer.trainable = False
for layer in base.layers[136:]:
    layer.trainable = True
```

```python
# Output braches with FLC layers
x = base.output # base model
x = GlobalAveragePooling2D()(x) # Global Pool before dividing into branches

# First branch for gender prediction


# Second branch for age prediction
age = Dense(512, activation='relu')(x)
age = Dropout(0.5)(age)
age_out = Dense(1, activation='linear', name='age_out')(age)

# Create the model
model = Model(inputs=base.input, outputs=[age_out])

model.summary()
```

```python
# Add callbacks and optimiser 
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',  
    factor=0.5,          
    patience=5,          
    min_lr=0.00001,      
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_loss',  
    patience=10,         
    verbose=1,
    restore_best_weights=True
)
    
adam = Adam(learning_rate=0.001)
```

```python
model.compile(optimizer=adam,
               loss={'age_out': 'mse'},
               metrics={'age_out': 'mae'})

history = model.fit(
    train_generator,
    epochs=30,
    validation_data=val_generator,
    callbacks=[early_stopping, reduce_lr]
)
```

```python
# Evaluate on a test set
test_loss, test_age_mae = model.evaluate(test_generator)

print("Test Loss:", test_loss)
print("Test Age MAE:", test_age_mae)
```

```python
fig = plt.figure(figsize=(12, 6))

# Loss
fig.add_subplot(2, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.legend()
plt.grid(True)
plt.ylim([0, max(max(history.history['loss']), max(history.history['val_loss'])) + 10])
plt.title('Loss')
plt.xlabel('Epoch')


plt.tight_layout()
plt.show()
```
